```python
city_list=pd.read_csv("C://Users/Julie/dev/GW_Homework/worldcities.csv")
number_of_cities=500
selected_citites=city_list.sample(n=number_of_cities)
selected_citites.head()
```

```python
#selected_citites
lat_list=[]
temps_list=[]
humidity_list=[]
cloudiness_list=[]
ws_list=[]
counter=0
print("Beginning Data Retrieval")
print("-------------------------")
for index, row in selected_citites.iterrows():
    counter=counter+1
    city_name=row["City"]
    citys_country=row["Country"].upper()
    city_lat=row["Latitude"]
    lat_list.append(city_lat)
    city_lng=row["Longitude"]
    time.sleep(1)
    city_url="http://api.openweathermap.org/data/2.5/weather?lat="+str(city_lat)+"&lon="+str(city_lng)+ "&appid="+api_key+"&units="+units
    city_weather=req.get(city_url).json()
    temp_in_faren=city_weather["main"]["temp"]
    temps_list.append(temp_in_faren)
    humidity_list.append(city_weather["main"]["humidity"])
    cloudiness_list.append(city_weather["clouds"]["all"])
    ws_list.append(city_weather["wind"]["speed"])
    print("Processing Record %s of %s: %s, %s"%(counter, number_of_cities, city_name.title(), citys_country))
    print(city_url)
print("-------------------------")
print("Data Retrieval Complete")
print("-------------------------")
```

### Make a DF and export it to CSV

```python
country_list=[]
city_list_cap=[]
for country in selected_citites["Country"]:
    country_list.append(country.upper())
for city in selected_citites["City"]:
    city_list_cap.append(city.title())
```

```python
weather=[("City",city_list_cap),
         ("Country",country_list),
        ("Latitude",lat_list),
         ("Longitude",selected_citites["Longitude"]),
        ("Temperature (F)", temps_list),
        ("Humidity", humidity_list),
        ("Cloudiness", cloudiness_list),
        ("Wind Speed", ws_list)]
weather_df=pd.DataFrame.from_items(weather)
weather_df.head()
```

```python
weather_df.to_csv("weather.csv")
```
